```python
# 44. How Do You Get the Items Not Common to Both Series a and Series B?
```

```python
s1=pd.Series([1,2,3,4,5])
s2=pd.Series([4,5,6,7,8])
```

```python
inter=np.intersect1d(s1,s2)
inter
```

```python
union=np.union1d(s1,s2)
union
```

```python
pd.Series(np.setdiff1d(union,inter))
```
