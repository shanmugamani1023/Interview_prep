```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image=cv2.imread("images/1.jpeg")
```

```python
hist,bins=np.histogram(image.flatten(),256,[0,256],density=True)
```

```python
plt.plot(hist)
plt.title("Before Equalization")
```

```python
gray=cv2.equalizeHist(image)
```

```python
hist2,bin=np.histogram(gray,200,[100,200])
```

```python
plt.plot(hist2)
plt.title("After Equalization")
```

```python
cv2.imshow("gray",gray)
cv2.imshow("original",image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#Normalization
```

```python
norm=cv2.normalize(image,None,0,255,cv2.NORM_L1)
```

```python
cv2.imshow("norm",norm)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hist,bin=np.histogram(norm.flatten(),255,[0,255])
```

```python
plt.plot(hist)
```
